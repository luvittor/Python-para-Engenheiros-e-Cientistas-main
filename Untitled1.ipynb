```python

import pandas as pd

for i in range(1,6):
    file = "teste_" + str(i) + ".txt"
    
    with open(file) as f:
        linhas = f.readlines()
        f.close
        
        linhas = linhas[linhas.index("TESTE\n")+1:]
        
        file_csv = file + ".csv"
        
        with open(file_csv, 'w') as f:
            for j in linhas:
                f.write(j)
                
            f.close
            
        df = pd.read_csv(file_csv, engine = 'python')

df.head()
        
```

```python
import pandas as pd

for i in range(1,6):
    file = "teste_" + str(i) + ".txt"
    
    with open(file) as f:
        linhas = f.readlines()
        f.close
        
        skiprows = linhas.index("TESTE\n")+1
            
        df = pd.read_csv(file, skiprows = skiprows)

#df.head()
df
```

```python
df.to_csv("teste.csv")
```
